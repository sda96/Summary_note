## DKIW 피라미드

통계는 데이터(자료)의 과학으로 우리가 가진 목적에 맞는 적절한 데이터를 만들기 위하여 무엇을 측정할 것인가를 결정하는 것 굉장히 중요합니다.

자료, 정보, 지식, 지혜의 관계를 나타내는 방법론으로 DIKW 피라미드 이론이 존재합니다.

### 자료

자료(data) : 하나 이상의 객체들이 가지는 **양적, 질적인 수치** 에 대한 집합을 의미합니다.
- 질적 자료(qualitative data) : 수치로써 측정이 불가능한 자료를 질적자료라고 부르며 범주형 자료라고도 부릅니다. ex) 성별, 혈액형, etc...
    - 명목형 자료
    - 순서형 자료
- 양적 자료(quantattive data) : 수치로써 측정이 가능한 자료를 양적자료라고 부릅니다. ex) 가격, 온도, 지수, etc...
    - 이산형 자료 : 셀 수 있는 양적 자료
    - 연속형 자료 : 셀 수 없는 양적 자료

### 정보

정보(information) : **자료를 특정한 목적의 의사결정을 위해 가공한 형태**를 정보라고 부르며, 좋은 정보는 목적 적합성과 신뢰성 및 적시성을 유지해야 합니다.    
정보의 수학적 정의 : 정보는 상대방에게 사건을 알릴때 전달되는 내용을 의미합니다. 예를 들어 상대방에게 자신의 집에 불이났다라는 정보를 전달할 때 필요한 정보량은 1비트 입니다. 

왜냐하면 "집에 불이 났다." 또는 "집에 불이 않났다."로 2가지 경우의 정보만 전달하면 되기 때문에 0과 1로 표현가능하므로 정보량은 1비트를 할당하게 됩니다. 사건이 발생할 확률의 분포가 균일분포라고 할 때, 정보량의 식은 아래와 같이 나옵니다. $$ I(x) = \log{\frac{1}{p(x)}} = -\log{p(x)}, 단, 0 \leq p(x) \leq 1 $$
<img src="https://imgur.com/mYl5Rbh.jpg">
발생가능 확률이 낮을수록 정보량은 커지게 됩니다. 이 의미는 가치있는 정보의 발생 확률은 작다는 뜻입니다. 그리고 발생가능 확률이 높을수록 정보량은 커지게 되며 이 말은 자주 발생하는 흔한 사건일수록 정보의 가치는 떨어지게 된다는 말입니다.

### 지식

지식(knowledge) : 정보들을 활용하여 목적의 패턴을 이해하고 이에 대한 예측의 결과물을 의미합니다.  

### 지혜

지혜(wisdom) : 지혜는 지식에 유연성을 더하여 상황이나 맥락에 맞게 지식을 적용하는 것을 의미합니다. 

### 데이터를 다룰때의 가져야 하는 태도

단순히 데이터의 양이 많다고 좋은 것이 아닌 무수한 데이터의 바다속에서 목적에 맞는 데이터를 찾아 의미를 부여하여 데이터를 정보로 만드는 과정이 중요합니다. 목적 적합성을 지닌 정보는 다른 정보들과 상호보완적으로 조합되어 지식이라는 시스템을 만들게 됩니다. 하나의 시스템인 지식은 다른 환경에도 적용되어 적용된 환경에 맞게 보완되어 보편적으로 사용되는 지식인 지혜가 됩니다.

그러므로, 통계를 한다는 것은 데이터를 통하여 목적에 맞는 정보를 도출하고, 정보들을 이용하여 만든 모델를 이용하여 **목적을 달성하는 도구**, 혹은 중요사항에 대한 **의사결정의 도구**로써 사용하는 것이 좋다고 생각합니다.

## 개체와 변수
개체(object) : 자료를 **설명하는 대상**으로 사람, 동물, 혹은 물체, 단순 수치가 될 수 있다.  
변수(variable) : 개체의 **어떤 특징**으로 개체마다 다른 값을 가질 수 있다.
- 리서치 분야 : 개체가 가지는 특징을 속성(attribute)라고 부르고 이러한 속성의 집합을 변수라고 부릅니다.
- 컴퓨터과학 분야 : 데이터가 저장되어 있는 메모리 주소명을 변수라고 부릅니다. 

## 모집단과 표본

모집단 : 정보를 얻기 워나는 개체의 전체 집단을 의미합니다.  
표본 : 모집단의 일부분으로서 전체에 관한 결론을 얻기 위햇 실제로 정보를 수집하게 되는 집단으로 모집단의 부분집합을 의미합니다.  
모수 : 모집단의 크기를 나타내는 값을 모수라고 부릅니다.  
통계량 : 표본집단의 특징을 나타내는 값을 통계량이라고 부릅니다.

## 편향와 분산

### 편향(bias)

편향 : 표본을 여러번 선택시 모수에서 한 방향으로 일관되고 반복적으로 벗어나는 통계량의 편차를 편의라고 부릅니다. 
추정량에서의 편향을 수식으로 말하자면 추정량과 실측값사이의 차이에 기댓값을 취한 식을 말합니다.
$$ Bias(\hat{f}) = E(\hat{\theta} - \theta) $$
- 편의를 줄이는 방법은 뽑는 표본들을 무작위(나올 수 있는 경우의 수들의 확률이 모두 동등할 때)로 뽑아야 합니다
- 편의를 줄이는 방법은 뽑는 표본들을 무작위(뽑는 표본의 분포가 균일분포가 되도록)로 뽑아야 합니다
- 편향은 **학습모델에서 잘못된 가정**을 하였을 때 발생하는 오차로 **과소적합 문제**를 발생시킵니다.

### 분산(variance)

분산 : 표본이 여러번 선택 시 통계량의 값이 기준(보통 평균)으로부터 얼마나 퍼져있는 가(거리가 얼마나 먼 가)를 의미합니다.
- 표본의 크기를 늘리면 분산이 작아집니다.
- 분산은 훈련 셋에 내재된 작은 변동 때문에 발생하는 오차입니다.
- 높은 분산값은 큰 노이즈까지 모델링에 포함시키기에 **과대적합 문제**를 발생시킵니다.

### 편향와 분산의 trade-off

훈련 데이터의 범위를 넘어 과대적합을 예방하기 위해서 두 종류
의 오차(편향, 분산)을 최소화할 때 겪는 문제입니다.  

모델 선택시 훈련 데이터의 패턴을 잘 파악하여 테스트 데이터에서도 예측, 분류를 잘하는 일반화된 모델을 만드는 것이 이상적이지만 현실적으로는 그러한 모델을 만드는 데 어려움이 발생합니다.

고분산 학습 모델은 학습 데이터의 분산을 모두 표현 가능한 모델은 훈련 데이터에서는 잘표현하지만 지나치게 큰 노이즈나 아예 부적절한 트레이닝 데이터까지 과적합의 위험이 있습니다.

### MSE(오차제곱평균)의 편향과 분산 분해 

표본공간의 원소인 x에 대한 **오차**는 모수 $\theta$와 추정량 $\hat{\theta}$의 차이입니다.
$$ \epsilon_{\hat{\theta}}(x) = \hat{\theta}(x) - \theta$$
**편향**은 그 오차의 기댓값입니다.
$$ B(\hat{\theta}) = E(\hat{\theta}(X) - \theta)$$
그리고 모수 $\theta$의 **불편추정량**은 편향이 0이되는 추정량을 말합니다.
$$ E(\hat{\theta}(X)) = \theta $$
**평균제곱오차**는 오차에 제곱을 한 기댓값을 말합니다.
$$ MSE(\hat{\theta}) = E[(\hat{\theta}(X) - \theta)^2]$$
평균제곱오차는 편향과 분산의 균형을 이루는 정량적인 기준이 되는 척도(손실함수)로써 사용되어 집니다. 그러므로 만들어진 모델들 중에서 평균제곱오차가 가장 작은 모델은 가장 잘만들어진, 일반화가 잘 되어진 적합한 모델이라고 말할 수 있습니다

평균제곱오차는 편향과 분산으로 분해가 가능합니다.
$$ 평균제곱오차 = 편향^2 + 분산 + 노이즈$$
분해 유도 공식은 참고사이트에 있습니다.

## 실험설계법

통계적 모델이나, 수학적 모델을 세우는 등의 실험설계를 위하여 사용되는 변수의 종류로는 **독립변수**(=설명변수)와 **종속변수**(=반응변수)가 존재합니다. 독립변수는 실험에서 나타난 효과의 **원인**을 제공하는 변수이고 종속변수는 독립변수에 의하여 변화된 **결과**를 나타내는 변수입니다.

모델을 세우는데 주의할 점은 **교락**과 **잠재변수**라는 개념입니다.  

### 교락

교락은 2개 이상의 독립변수에 의해 변화되어진 종속변수가 **어떤 독립변수에 의해서 변화되어진지 구별할 수 없는 상태**를 말합니다.  

### 잠재변수

잠재변수는 **실험설계에 포함되어지지 않았지만 종속변수에는 영향을 끼치는 변수**를 말하며, 잠재변수가 교락되어질 수도 있습니다.

## 상관분석

### 상관분석

상관분석은 모델에서 두 변수의 관계가 서로 선형인지 비선형인지 알기 위한 분석을 말합니다. 상관분석의 수치상 결과인 상관계수는 -1부터 1사이의 값을 가지며 0에 가까울수록 두 변수는 비선형적 관계를 가집니다.   

상관계수를 구하는 방법으로는 피어슨 상관분석, 스피어만 상관분석, 켄달타우 검정이 있습니다.  
- 피어슨 상관분석은 모수의 특정 분포를 가정하여 접근하는 방법인 모수적 방법입니다.(여기서 두 변수의 분포는 정규분포여야 합니다.)
- 스피어만 상관분석과 켄달 타우 검정은 모수의 분포를 가정하지 않고 접근하는 비모수적 방법입니다.

스피어만 상관분석과 켄달 타우 검정과 같은 비모수적 방법은 데이터의 개수가 적은 경우에 자주 사용되어 집니다.

### 선형성이란

함수 $f$(또는 연산)에 대해,

- 가산성(Additivity), 즉, 임의의 수 $x$, $y$에 대해 $f(x+y)=f(x)+f(y)$가 항상 성립하고,
- 동차성(Homogeneity), 즉, 임의의 수 $x$와 $\alpha$ 에 대해 $f(\alpha x)=\alpha f(x)$가 항상 성립할 때,
- 두 식을 동시에 만족하는 식은, 임의의 수 x,y에 대해 $f(a_1x + a_2y) = a_1f(x) + a_2f(y)$입니다.

함수 $f$는 선형성을 가집니다.

마지막 식의 형태는 행렬에서는 두 행렬의 선형결합 형태와 동일하고 벡터에서는 두벡터의 내적과정과 동일합니다.  
그러므로 두 행렬 또는 두 벡터는 서로 선형성을 지니고 있다고 볼 수 있습니다.

선형성이 강하다는 것은 변수들의 관계가 직선에 의해 잘 모델링(선형성의 조건을 충족하는 함수)된다는 뜻입니다.

### 피어슨 상관분석

두 양적 변수인 $X$와 $Y$의 상관계수를 구하는 분석으로 두 변수의 공분산에 각 변수의 표준편차 곱한값을 나누어주면 나오는 식으로 식은 아래와 같습니다.

$$ r_{XY} = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{\sqrt{\sum{(X_i - \bar{X})^2}}\sqrt{\sum{(Y_i - \bar{Y})^2}}}$$
상관계수 $r_{XY}$는 -1에서 1사이의 값을 가집니다.  
상관계수의 식에서 사용된 두 변수사이의 공분산 덕분에 두 변수의 선형관계를 알 수가 있습니다.  
- $r_{XY}$의 값이 -1에서 0사이의 값이면 두 변수는 **음의 상관관계**를 가지며 두 변수중 하나의 변수가 증가하면 다른 변수는 감소하는 경향을 보입니다.  
- $r_{XY}$의 값이 1에서 0사이의 값이면 두 변수는 **양의 상관관계**를 가지며 두 변수중 하나의 변수가 증가하면 다른 변수도 증가하는 경향을 보입니다.  
- $r_{XY}$의 값이 0이면 두 변수는 선형성을 가지지 않는 비선형을 가지며 공분산도 0이 되기 때문에 두 변수 $X,Y$는 **서로 독립의 관계**를 가집니다.

**상관계수값이 높을수록 선형성이 높아지고 두 변수의 관련성은 깊어집니다.**

### 스피어만 순위상관분석

피어슨은 선형관계를 활용한 분석이라면, 스피어만의 순위상관분석은 두 변수의 단조관계를 활용한 분석입니다. 순위상관계수가 높으면 두 변수의 순서는 서로 비슷하다는 의미이고 낮으면 두 변수의 순서는 전혀 다르다는 의미가 됩니다.  
스피어만의 순위상관분석에 사용하기에 적당한 변수는 이산형 변수와 순서를 가지고 있는 순서형 질적변수가 있습니다.

$$r_s = 1 - \frac{6\sum{d_{i}^2}}{n(n^2-1)}$$
$d_{i}$는 두 변수 집합에서 동일한 순서를 가진 원소끼리 뺀 값입니다.  
$n$은 표본의 개수를 의미합니다.
- 스피어만 상관계수는 1에 가까울수록 한 변수가 커지면 다른변수도 단조성을 가지고 커지게 됩니다.
- 스피어만 상관계수가 0에 가까울수록 단조성을 띄지 않게 됩니다.

스피어만은 데이터 내 편차와 에러에 민감하며 일반적으로 켄달 상관계수보다 높은 값을 가진다.

로그함수는 가산성을 만족하지 못하므로 비선형입니다.  
지수함수는 가산성을 만족하지 못하므로 비선형입니다.  
원점을 지나지 않는 직선은 가산성을 만족하지 못하므로 비선형입니다.

### 상관관계 해석시 주의할 점

상관분석을 해석하는데 주의할 점은 상관계수가 높다고 두 변수가 인과관계를 가지는 것은 아닙니다.  
예를 들어, 코스피 지수와 사람들의 평균키의 상관관계를 분석한 결과 이 둘은 양의 상관관계를 가진다고 가정합니다. 두 변수가 서로 양의 상관관계라고 코스피 지수가 높기 때문에 사람들의 평균키가 커지고 있다는 인과관계로 이해하면 안됩니다.  

상관관계는 단순히 두 변수가 관련성이 있다는 것을 알려줄 뿐이지 두 변수의 인과관계를 표현한 것은 아닙니다. 두 변수의 인과관계를 알기 위해서는 회귀분석을 통하여 나온 모델로 수치적인 인과관계를 알 수 있습니다.
